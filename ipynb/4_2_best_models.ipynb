```python
import pandas as pd

DATA_LOC = '../../../Data/microbiome_xai/'
FILE_LOC = DATA_LOC + "micom-sim-data/ml_params_fd_PARALLEL.csv"
```

```python
param_df = pd.read_csv(FILE_LOC)
```

```python
new_col_ids = ['phenotype', 'input_type', 'reg_val', 'n_features', 'reg_type',
       'feat_type', 'norm_type', 'accuracy', 'balanced_accuracy', 'roc_auc',
       'average_precision', 'f1']
param_df.columns = new_col_ids
param_df.head()
```

```python
N_TOP_PARAMS = 5
FEATS_LIMIT = 75

param_df = param_df[param_df["n_features"]<FEATS_LIMIT].copy()
param_df = param_df[param_df["norm_type"]=="StandardScaler"].copy()
view_cols = ['reg_val', 'n_features', 'reg_type', 'norm_type', "roc_auc", 'balanced_accuracy',"average_precision", "f1"]
top_params_df = pd.DataFrame()
for pheno_input, grp in param_df.groupby(["phenotype", "input_type"]):
    print(pheno_input)
    # print(grp.sort_values(["roc_auc"],ascending=False)[:N_TOP_PARAMS][view_cols])

    pheno_input_top_df = grp.sort_values(["roc_auc"],ascending=False)[:N_TOP_PARAMS][view_cols].copy()
    # pheno_input_top_df = grp.sort_values(["balanced_accuracy"],ascending=False)[:N_TOP_PARAMS][view_cols].copy()
    # pheno_input_top_df = grp.sort_values(["f1"],ascending=False)[:N_TOP_PARAMS][view_cols].copy()
    pheno_input_top_df["pheno_input"] = "_".join(pheno_input)
    top_params_df = pd.concat([top_params_df, pheno_input_top_df],axis=0)

top_params_df.head()
```
